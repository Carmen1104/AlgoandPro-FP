````python
dff = dff._convert(numeric=True)
dff['BEGIN_MONTH'] =dff['BEGIN_MONTH'].astype(float)
dff = dff.replace(np.nan, 0, regex=True)
n = 'dataCleaned.csv'
dff.to_csv(n, index=False)
dff.info()
````
